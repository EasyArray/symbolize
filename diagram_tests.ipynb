```python
from symbolize.relation.diagram import *
```

```python
leaf1 = pred("over", 2, "ext1")
leaf2 = pred("under", 2, "ext1")

id1, id2 = id("A"), id("B")

conj = op("&", leaf1, leaf2, value="True")

conj.render()
```

```python
neg = op("~", conj, value="False")
neg.render()
```

```python
dia = app(conj, id1, id2, value="True")
dia.render()
```

```python
dia = app(neg, id1, id2, value="False")
dia.render()
```
